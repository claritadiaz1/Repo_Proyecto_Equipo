```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("IPG2022_limpio.csv")
```

```python
macro = (
    df.groupby("macrozona")
      .agg({
          "area_km2": "sum",
          "codigo_glaciar": "count"
      })
      .reset_index()
)

macro.columns = ["macrozona", "superficie_total", "cantidad_glaciares"]

macro
```

```python
import altair as alt

grafico = alt.Chart(macro).mark_bar().encode(
    x=alt.X(
        "superficie_total:Q",
        title="Superficie glaciar total (km²)"
    ),

    y=alt.Y(
        "macrozona:N",
        sort="-x",
        title="Macrozona"
    ),

    tooltip=[
        "macrozona",
        "superficie_total",
        "cantidad_glaciares"
    ]
).properties(
    width=700,
    height=400,
    title="Superficie glaciar por macrozona en Chile"
)

grafico
```

```python
grafico.save("vis_01.html")
```
