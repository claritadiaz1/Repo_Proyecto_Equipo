```python
import pandas as pd
import altair as alt

# cargar datos
df = pd.read_csv('IPG2022_limpio.csv')

# top 10 cuencas con mas agua para procesar
top_cuencas = df.groupby('cuenca')['volumen_agua_km3'].sum().reset_index()
top_cuencas = top_cuencas.sort_values('volumen_agua_km3', ascending=False).head(10)

# grafico horizontales
chart = alt.Chart(top_cuencas).mark_bar(color='#4682b4').encode(
    x=alt.X('volumen_agua_km3:Q', title='Volumen Agua (km³)'),
    y=alt.Y('cuenca:N', sort='-x', title='Cuenca'),
    tooltip=['cuenca', 'volumen_agua_km3']
).properties(title='Top 10 Cuencas con Reservas de Agua Glaciar', width=500)

# guardar grafico
chart.save('vis_01.html')
```
